```python
import numpy as np
from scipy import signal, fftpack
import matplotlib.pyplot as plt
```

```python
Fs = 10000
freq_Nyq = Fs/2
[B,A] = signal.butter(4, 1000/freq_Nyq, 'lowpass')
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.butter.html

w, H = signal.freqz(B, A)
f= freq_Nyq*(w/np.pi)
plt.plot(f, abs(H))

plt.title('Magnitude da resposta em frequência de um filtro Butterworth')
plt.xlabel('frequência [Hz]')
plt.ylabel('Magnitude')
plt.grid()
plt.show()

angles = np.unwrap(np.angle(H))
plt.plot(f, angles, 'g')
plt.ylabel('Angle (radians)', color='g')
plt.xlabel('frequência [Hz]')
plt.grid()
plt.show()
```

#Função Impulso discreto $\delta[n]$:


```python
Fs = 2000 # taxa de amostragem
x = signal.unit_impulse(32,0)
n = np.linspace(0,31,32) # np.linspace(início, fim , número de amostras)
plt.figure()
plt.grid()
plt.stem(n,x)
plt.title('Função impulso iniciando em n=0')
```

```python
# DFT através do scipy
N = len(x)
DFT = fftpack.fftshift(fftpack.fft(x,N))
mag_espectro = np.abs(DFT)
f = fftpack.fftshift(fftpack.fftfreq(N, 1 / Fs))
plt.stem(f,mag_espectro)
plt.grid()
plt.title('Espectro de magnitude em frequência')
```

```python
resp_fase = np.unwrap(np.angle(DFT))
plt.plot(f, resp_fase, 'g')
plt.ylabel('Angle (radians)', color='g')
plt.xlabel('frequência [Hz]')
plt.grid()
plt.show()
```

Alterando a fase inicial:

```python
Fs = 2000 # taxa de amostragem
f0 = 100 # frequência da senoide
fase_ini = np.pi/2 # ângulo de fase inicial
ampl = 1 # amplitude da senoide
durt = 0.1 # duração do sinal
Ts = 1/Fs # Periodo de amostragem
num_passos = np.round(durt*Fs).astype(int) # número de passos
n = np.arange(0,num_passos,1) # índices amostrais
x_sine2 =  ampl*np.sin(2*np.pi * f0 * n*Ts + fase_ini)
t = n*Ts # escala de tempo do sinal
plt.plot(t, x_sine2)
```

```python
Fs = 4000 # taxa de amostragem
f0 = 500 # frequência da senoide
fase_ini = 0 # ângulo de fase inicial
ampl = 1 # amplitude da senoide
durt = 5 # duração do sinal
Ts = 1/Fs # Periodo de amostragem
num_passos = np.round(durt*Fs).astype(int) # número de passos
n = np.arange(0,num_passos,1) # índices amostrais
x_sinepuro =  ampl*np.sin(2*np.pi * f0 * n*Ts + fase_ini)
noise = np.random.normal(0,2,len(x_sinepuro))
t = n*Ts # escala de tempo do sinal
plt.plot(t, x_sinepuro)
```

```python

plt.title('Plot do sinal senoidal + ruido')
plt.plot(t, x_sinepuro+noise)
plt.ylabel('Magnitude')
plt.xlabel('Tempo (s)')
plt.grid()
plt.show()
```

```python
b, a = signal.cheby2(5, 20, [480, 520], btype='bandpass', fs=Fs)
w, h = signal.freqz(b, a, fs=Fs)
plt.plot(w, abs(h))
plt.title('Resposta em Frequência do Filtro')
plt.xlabel('Frequency [radians / second]')
```

```python
y_bandpass = signal.lfilter(b, a, x_sinepuro+noise)

plt.title('Plot do sinal senoidal filtrado')
plt.plot(t, y_bandpass)
plt.ylabel('Magnitude')
plt.xlabel('Tempo (s)')
plt.grid()
plt.show()
```

```python
# DFT of y with pure sine
DFT = fftpack.fft(x_sinepuro, len(x_sinepuro) )
mag_espectro = np.abs(DFT)
f = fftpack.fftfreq( len(x_sinepuro), 1/Fs )
plt.stem(f,mag_espectro)
plt.title('Espectro de magnitude da senoide pura')
plt.ylabel('Magnitude')
plt.xlabel('Frequência (Hz)')
plt.grid()
plt.show()
```

```python
# DFT of y with noisy sine
DFT = fftpack.fft( x_sinepuro+noise, len(x_sinepuro+noise) )
mag_espectro = np.abs(DFT)
f = fftpack.fftfreq( len(x_sinepuro+noise), 1/Fs )
plt.stem(f,mag_espectro)
plt.title('Espectro de magnitude da senoide com ruido')
plt.ylabel('Magnitude')
plt.xlabel('Frequência (Hz)')
plt.grid()
plt.show()
```

```python
DFT = fftpack.fft( y_bandpass, len(y_bandpass) )
mag_espectro = np.abs(DFT)
f = fftpack.fftfreq( len(y_bandpass), 1/Fs )
plt.stem(f,mag_espectro)
plt.title('Espectro de magnitude filtrada')
plt.ylabel('Magnitude')
plt.xlabel('Frequência (Hz)')
plt.grid()
plt.show()
```
