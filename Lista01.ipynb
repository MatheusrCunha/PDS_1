```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for n
n = np.arange(-5, 6)  # Adjust the range as needed to visualize

# Define the signal x1[n]
x1 = np.zeros_like(n)
x1[n == -1] = 1
x1[n == 1] = -1

# Plot the signal
plt.stem(n, x1, basefmt='b', use_line_collection=True)
plt.xlabel('n')
plt.ylabel('x1[n]')
plt.title('Discrete Signal x1[n] = δ[n + 1] - δ[n - 1]')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for n
n = np.arange(-5, 6)  # Adjust the range as needed to visualize

# Define the signal x2[n]
x2 = np.zeros_like(n)
x2[n == 0] = 1
x2[n == 2] = -2

# Plot the signal
plt.stem(n, x2, basefmt='b', use_line_collection=True)
plt.xlabel('n')
plt.ylabel('x2[n]')
plt.title('Discrete Signal x2[n] = δ[n] - 2δ[n - 2]')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Vetor de valores de n
n = np.arange(0, 10)  # Altere o intervalo de acordo com suas necessidades

# Defina o sinal x1[n]
x1 = np.zeros_like(n)
x1[np.where(n == -1)] = 1
x1[np.where(n == 1)] = -1

# Defina o sinal x2[n]
x2 = np.zeros_like(n)
x2[np.where(n == 0)] = 1
x2[np.where(n == 2)] = -2

# Inicialize a saída da convolução y[n] como zeros
y = np.zeros_like(n)

# Calcule a convolução
for n_index in range(len(n)):
    y[n_index] = np.sum(x1 * np.roll(x2, n_index))

# Plote o resultado
plt.stem(n, y, use_line_collection=True)
plt.xlabel('n')
plt.ylabel('y[n]')
plt.title('Convolução x1[n] * x2[n]')
plt.grid(True)
plt.show()



```
